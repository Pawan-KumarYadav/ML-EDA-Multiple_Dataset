```python
x_encoded =pd.get_dummies(X , columns= ['model', 'transmission', 'fuelType'],drop_first=True)
```

```python
x_encoded =pd.get_dummies(X)
x_encoded =x_encoded.astype(int)
```

```python
x_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
encode =LabelEncoder()
columns =['model' , 'transmission', 'fuelType']

Xlable = X

for i in columns:
    Xlable[i] = encode.fit_transform(Xlable[i])




     
```

```python
Xlable
```

```python
Xlable['model'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_col =['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
x_encoded[numerical_col] =scaler.fit_transform(x_encoded[numerical_col])
```

```python
x_encoded
```

```python
Xlable.columns
```

```python
Xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] =scaler.fit_transform(Xlable[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] )
```

```python
Xlable
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_encoded, y,
    test_size=0.33,
    random_state=42
)
```

```python
model =LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred =model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1-((1-r2) *(n-1)/ (n-p-1))
print(" adjusted r2 : " , adjusted_r2)
```
